```python
import matplotlib.pyplot as plt
import json
```

```python
def make_experiment_list(folder, num_heads_list,num_encoder_layers_list,dim_feedforward_encoder_list,dropout_list, lr = 0.0005, bz = 8):
	experiments_results = [{"num_heads" : num_heads, "num_encoder_layers":num_encoder_layers, "dim_feedforward_encoder": dim_feedforward_encoder, "dropout": dropout } for num_heads in num_heads_list for num_encoder_layers in num_encoder_layers_list for dim_feedforward_encoder in dim_feedforward_encoder_list for dropout in dropout_list]

	for exp in experiments_results:
		num_heads = exp["num_heads"]
		num_encoder_layers = exp["num_encoder_layers"]
		dim_feedforward_encoder = exp["dim_feedforward_encoder"]
		dropout = exp["dropout"]

		exp1 = f"../exps/{folder}/models/classification/traindata_bslcp/i3d_kinetics_bslcp_981/supervised/{num_heads}_{num_encoder_layers}_{dim_feedforward_encoder}_{dropout}_"+str(bz)+"_"+str(lr)+"_weighted_opt/seed_0/"
		exp1_train = exp1 + "train_results.json"
		exp1_eval = exp1 + "eval_results.json"


		exp["train_results"] = json.load(open(exp1_train))
		exp["eval_results"] = json.load(open(exp1_eval))

	return experiments_results
```

```python
def plot_experiment_results(experiment_results):
	for exp in experiments_results:
		num_heads = exp["num_heads"]
		num_encoder_layers = exp["num_encoder_layers"]
		dim_feedforward_encoder = exp["dim_feedforward_encoder"]
		dropout = exp["dropout"]
		train_results = exp["train_results"]
		eval_results = exp["eval_results"] 
		
		print("Number of heads" ,str(num_heads))
		print("Number of layers" , str(num_encoder_layers))
		print("Dim feedforward encoder" ,str(dim_feedforward_encoder))
		print("Dropout" , str(dropout))
		print()

		print("Max mF1B:", max([eval_results[k]["mF1B"] for k in train_results.keys()]))
		
		plt.figure(figsize=(8, 6), dpi=80)
		plt.title("mF1B  - " +  str(num_heads) + "," + str(num_encoder_layers) +"," + str(dim_feedforward_encoder) + "," + str(dropout))
		plt.plot([train_results[k]["mF1B"] for k in eval_results.keys()], label ="Train")
		plt.plot([eval_results[k]["mF1B"] for k in train_results.keys()], label ="Eval")
		plt.legend()
		plt.show()



		plt.figure(figsize=(8, 6), dpi=80)
		plt.title("mF1S  - " +  str(num_heads) + "," + str(num_encoder_layers) +"," + str(dim_feedforward_encoder) + "," + str(dropout))
		plt.plot([train_results[k]["mF1S"] for k in eval_results.keys()], label ="Train")
		plt.plot([eval_results[k]["mF1S"] for k in train_results.keys()], label ="Eval")
		plt.legend()
		plt.show()

		print("--------------------------------------------------------------------")
```
